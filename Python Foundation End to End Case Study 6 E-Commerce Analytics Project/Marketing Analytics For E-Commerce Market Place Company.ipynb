### b. Understanding how many new customers acquired every month

```python
Cust_Orders=pd.merge(left=Customers,right=Orders,on="customer_id", indicator = True)
Cust_Orders.head()
```

```python
Cust_Orders.order_purchase_timestamp=pd.to_datetime(Cust_Orders.order_purchase_timestamp,format='%m/%d/%Y %H:%M')
```

```python
Cust_Orders['Yr_Mth']=Cust_Orders.order_purchase_timestamp.apply(lambda x:pd.Timestamp.strftime(x,format="%Y-%m"))
Cust_Orders.head()
```

```python
Cust_YrMth=Cust_Orders.loc[:,['customer_unique_id','Yr_Mth']]
```

```python
Cust_YrMth.drop_duplicates(subset=['customer_unique_id'],keep='first',inplace=True)
```

```python
CustAcquired=Cust_YrMth.groupby('Yr_Mth')['customer_unique_id'].count()
CustAcquired
```

### c. Understand the retention of customers on month on month basis

```python
Cust_PrMth_Cnt=Cust_Orders.groupby('Yr_Mth')[['customer_unique_id']].count().reset_index()
Cust_PrMth_Cnt
```

```python
Cust_PrMth_Cnt.customer_unique_id.sum()
```

```python
Cust_Retn=pd.merge(left=Cust_PrMth_Cnt,right=CustAcquired,on='Yr_Mth')
Cust_Retn.head()
```

```python
Cust_Retn['Customers_Retained']=Cust_Retn.customer_unique_id_x - Cust_Retn.customer_unique_id_y
Cust_Retn
```

```python
Tot_Rev=Cust_Ord_Pay.groupby('Yr_Mth')[['Total_Payment']].sum()
Tot_Rev
```

### 2. Performing Customers/sellers Segmentation

### a. Divide the customers into groups based on the revenue generated 

```python
Customer_Segment=COOIPPS.groupby('customer_unique_id')[['Total_Payment']].sum().rename(columns={'Total_Payment':'Amount_Spent'})
```

```python
Customer_Segment['Cust_Grp']=np.where(Customer_Segment.Amount_Spent<2000,'Low_Revenue_Cust',np.where(Customer_Segment.Amount_Spent>4000,'High_Rev_Cust','Mid_Rev_Cust'))
Customer_Segment
```

### b. Divide the sellers into groups based on the revenue generated 

```python
Seller_Segment=COOIPPS.groupby('seller_id')[['price']].sum().rename(columns={'price':'Amount_Sold'})
```

```python
Seller_Segment['Seller_Grp']=np.where(Seller_Segment.Amount_Sold<2000,'Low_Revenue_Seller',np.where(Seller_Segment.Amount_Sold>3500,'High_Revenue_Seller','Mid_Revenue_Seller'))
Seller_Segment
```
